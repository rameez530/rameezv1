```python
data=data.drop(['Loan_ID'],axis=1)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
ob1=LabelEncoder()
ob2=LabelEncoder()
ob3=LabelEncoder()
ob4=LabelEncoder()
ob5=LabelEncoder()
ob6=LabelEncoder()
data['Gender']=ob1.fit_transform(data['Gender'])
data['Married']=ob2.fit_transform(data['Married'])
data['Dependents']=ob3.fit_transform(data['Dependents'])
data['Education']=ob4.fit_transform(data['Education'])
data['Self_Employed']=ob5.fit_transform(data['Self_Employed'])
data['Property_Area']=ob6.fit_transform(data['Property_Area'])
data
```

```python
data.corr()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(data.select_dtypes(['number']).corr(),annot=True)
```

```python
x=data.drop(['Loan_Amount_Term'],axis=1)
y=data['Loan_Amount_Term']
x
```

```python
from sklearn.feature_selection import SelectKBest
bestfeature=SelectKBest()
fit=bestfeature.fit(x,y)
fit
```

```python
dfscore=pd.DataFrame(fit.scores_,columns=['score'])
dfcol=pd.DataFrame(x.columns,columns=['feature'])
fs=pd.concat([dfcol,dfscore],axis=1)
fs
```

**LOGISTIC REGRESSION**

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtest
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
print(model1.score(xtest,ytest))
print(model1.score(xtrain,ytrain))
```

```python
y_pred1=model1.predict(xtest)
y_pred1
```

```python
model1.predict([[0,1,2,0,0,2779,3664,360.0,0.0,1]])
```
